```python
%matplotlib widget

from utils import visualization, weights, features, build, embedding, evaluation
from sklearn.datasets import load_digits
from data import preprocessing, small
import networkx as nx
from sklearn.manifold import trustworthiness
```

```python
digits = load_digits()
data = digits.data
labels = digits.target
data, labels = preprocessing.remove_duplicities(data, labels, shuffle=True, normalize=True)
```

# define builders

```python
cheapest = build.CheapestBuilder(weight_fun=weights.reciprocal)
cheapest.build(data)
```

```python

spanning = build.SpanningTreeBuilder(weight_fun=weights.reciprocal)
spanning.build(data)
```

```python

nn_cheapest_2 = build.CheapestNNBuilder(weight_fun=weights.reciprocal, knn=2)
nn_cheapest_2.build(data)
```

```python

nn_spanning_2 = build.SpanningNNBuilder(weight_fun=weights.reciprocal, knn=2)
nn_spanning_2.build(data)
```

```python

hierarchical = build.HierarchicalBuilder(weight_fun=weights.reciprocal, knn=1)
hierarchical.build(data)
```

```python

cluster = build.HierarchicalClusterBuilder(weight_fun=weights.reciprocal, knn=2)
cluster.build(data)
```

```python
nx.algorithms.isomorphism.is_isomorphic(spanning.graph, hierarchical.graph, edge_match=nx.algorithms.isomorphism.numerical_edge_match('weight', 1.0))
```

# to 2D

```python
embedder = embedding.KamadaKawaiEmbedder(embedding_dim=2)
```

```python
embedder.embed(cheapest.graph)
cheapest_emb = embedder.embeddings

evaluation.print_evaluation(data, cheapest_emb)
visualization.show_data(cheapest_emb, labels=labels, square=True, title='Cheapest build', dpi=500, outpath='../kamada/cheapest_2D.png')
```

```python
embedder.embed(spanning.graph)
spanning_emb = embedder.embeddings

evaluation.print_evaluation(data, spanning_emb)
visualization.show_data(spanning_emb, labels=labels, square=True, title='Spanning tree build', dpi=500, outpath='../kamada/spanning_2D.png')
```

```python
embedder.embed(nn_cheapest_2.graph)
nn_cheapest_2_emb = embedder.embeddings

evaluation.print_evaluation(data, nn_cheapest_2_emb)
visualization.show_data(nn_cheapest_2_emb, labels=labels, square=True, title='2-NN cheapest build', dpi=500, outpath='../kamada/nn_cheapest_2D.png')
```

```python
embedder.embed(nn_spanning_2.graph)
nn_spanning_2_emb = embedder.embeddings

evaluation.print_evaluation(data, nn_spanning_2_emb)
visualization.show_data(nn_spanning_2_emb, labels=labels, square=True, title='2-NN spanning build', dpi=500, outpath='../kamada/nn_spanning_2D.png')
```

```python
embedder.embed(hierarchical.graph)
hierarchical_emb = embedder.embeddings

evaluation.print_evaluation(data, hierarchical_emb)
visualization.show_data(hierarchical_emb, labels=labels, square=True, title='Hierarchical build', dpi=500, outpath='../kamada/hierarchical_2nn_2D.png')
```

```python
embedder.embed(cluster.graph)
cluster_emb = embedder.embeddings

evaluation.print_evaluation(data, cluster_emb)
visualization.show_data(cluster_emb, labels=labels, square=True, title='Hierarchical cluster build', dpi=500, outpath='../kamada/cluster_2D.png')
```
